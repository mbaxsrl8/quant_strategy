# Showcase various factors in astrader library

```python
from atrader import *

factor_data_1 = get_factor_by_code(factor_list=['pe_ratio_ttm', 'pb_ratio_ttm'], target='SZSE.000001', begin_date='2025-01-01', end_date='2025-06-30')
print(factor_data_1)
```

```python
factor_data_3 = get_factor_by_factor(factor='pe_ratio_ttm', target_list=['SZSE.000001', 'SZSE.000002'], begin_date='2025-01-01', end_date='2025-06-30')
print(factor_data_3)
```
