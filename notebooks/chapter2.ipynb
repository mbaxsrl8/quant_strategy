# Minimal atrader context demo

This notebook runs a prepared strategy file and prints `context.account_list` and `context.target_list` from inside `init(context)`.

```python
from pathlib import Path

from atrader import run_backtest

project_root = next(
    path for path in [Path.cwd(), *Path.cwd().parents]
    if (path / "requirements.txt").exists()
 )
strategy_file = project_root / "notebooks" / "print_context.py"

print(strategy_file)

result = run_backtest(
    strategy_name="print_context_lists",
    file_path=str(strategy_file),
    target_list=["SHFE.RB0000"],
    frequency="day",
    fre_num=15,
    begin_date="2025-03-01",
    end_date="2025-03-05",
)

print(result)
```
